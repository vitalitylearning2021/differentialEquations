```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python
# Define the ODE system (simple harmonic oscillator)
def harmonic_oscillator(t, y):
    dydt = [y[1], -y[0]]
    return dydt
```

```python
# Initial conditions: y1 (position) = 1, y2 (velocity) = 0
y0 = [1.0, 0.0]
t_span = (0, 10)  # Time range from 0 to 10
```

```python
# Solve the ODE using solve_ivp with the RK45 method
solution = solve_ivp(
    harmonic_oscillator, 
    t_span, 
    y0, 
    method='RK45', 
    t_eval=np.linspace(0, 10, 100)
)
```

```python
# Plotting the results
plt.figure(figsize=(10, 5))
plt.plot(solution.t, solution.y[0], label="$y_1$")
plt.plot(solution.t, solution.y[1], label="$y_2$")
plt.xlabel("$t$")
plt.ylabel("$y$")
#plt.title("Simple Harmonic Oscillator Solution Using Runge-Kutta (RK45)")
plt.legend()
plt.grid(True)
plt.show()
```
